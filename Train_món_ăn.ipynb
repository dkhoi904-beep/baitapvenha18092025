```python


# BƯỚC 1: CÀI ĐẶT THƯ VIỆN `split-folders`
# Thư viện này cực kỳ hữu ích, giúp chia dữ liệu theo cấu trúc thư mục một cách tự động.
!pip install split-folders
print("✅ Đã cài đặt thư viện split-folders!")


# BƯỚC 2: THỰC HIỆN CHIA DỮ LIỆU
import splitfolders
import os

# --- VUI LÒNG THAY ĐỔI CÁC ĐƯỜNG DẪN BÊN DƯỚI ---

# 1. Đường dẫn đến thư mục GỐC chứa 17 thư mục món ăn của bạn
# Ví dụ: '/content/drive/MyDrive/DuAnCuoiKy/DatasetMonAn'
input_folder = '/content/Data Project'

# 2. Đường dẫn đến thư mục bạn muốn LƯU kết quả sau khi chia
# Thư mục này sẽ được tự động tạo ra, không cần bạn tạo trước.
# Ví dụ: '/content/drive/MyDrive/DuAnCuoiKy/DatasetMonAn_Split'
output_folder = '/content/final dataset'

# ----------------------------------------------------


print(f"\n📁 Bắt đầu quá trình chia dữ liệu...")
print(f"   - Dữ liệu đầu vào: {input_folder}")
print(f"   - Dữ liệu đầu ra: {output_folder}")

# Kiểm tra xem đường dẫn đầu vào có tồn tại hay không để tránh lỗi
if not os.path.exists(input_folder):
    print(f"\n❌ LỖI: Đường dẫn '{input_folder}' không tồn tại. Vui lòng kiểm tra lại!")
else:
    # Thực hiện chia dữ liệu với tỷ lệ: 70% train, 15% validation, 15% test
    # seed=42 để đảm bảo mỗi lần bạn chạy lại code, kết quả chia vẫn y hệt.
    splitfolders.ratio(
        input_folder,
        output=output_folder,
        seed=42,
        ratio=(0.7, 0.15, 0.15),
        group_prefix=None
    )

    print(f"\n🎉 HOÀN TẤT! Dữ liệu đã được chia thành công.")
    print(f"   - Kết quả được lưu tại thư mục: {output_folder}")
    print("   - Bên trong sẽ có 3 thư mục con: 'train', 'val', và 'test'.")
```

```python


# Import các thư viện cần thiết
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

print("✅ TensorFlow version:", tf.__version__)
print("✅ Các thư viện đã được import!")

# BƯỚC 1: CHUẨN BỊ DỮ LIỆU
# --- VUI LÒNG THAY ĐỔI ĐƯỜNG DẪN BÊN DƯỚI ---

# Đường dẫn đến thư mục chứa 3 thư mục train, val, test mà bạn đã tạo ở bước trước
# Ví dụ: '/content/drive/MyDrive/DuAnCuoiKy/DatasetMonAn_Split'
base_dir = '/content/final dataset'

# ----------------------------------------------------

train_dir = f'{base_dir}/train'
validation_dir = f'{base_dir}/val'

# Thiết lập các tham số cho ảnh và batch
IMG_SIZE = (224, 224) # Kích thước ảnh đầu vào cho MobileNetV2
BATCH_SIZE = 32      # Số ảnh xử lý trong 1 lần

# Tạo các trình tạo dữ liệu (Data Generators)
# ImageDataGenerator sẽ tự động đọc ảnh từ các thư mục, gán nhãn,
# và thực hiện tăng cường dữ liệu (data augmentation) cho tập train.
train_datagen = ImageDataGenerator(
    rescale=1./255,         # Chuẩn hóa giá trị pixel về khoảng [0, 1]
    rotation_range=40,      # Xoay ảnh ngẫu nhiên trong khoảng 40 độ
    width_shift_range=0.2,  # Dịch chuyển ảnh theo chiều ngang
    height_shift_range=0.2, # Dịch chuyển ảnh theo chiều dọc
    shear_range=0.2,        # Bóp méo ảnh
    zoom_range=0.2,         # Phóng to/thu nhỏ ảnh
    horizontal_flip=True,   # Lật ảnh theo chiều ngang
    fill_mode='nearest'
)

# Với tập validation, ta chỉ cần rescale, không cần tăng cường dữ liệu
validation_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical' # Vì ta có nhiều lớp (17 món ăn)
)

validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

print("\n✅ Đã tạo xong các trình tạo dữ liệu (Data Generators)!")

# BƯỚC 2: XÂY DỰNG MÔ HÌNH (TRANSFER LEARNING)

# Tải mô hình MobileNetV2 đã được huấn luyện trước trên bộ dữ liệu ImageNet
# include_top=False: Bỏ đi lớp phân loại cuối cùng của mô hình gốc
base_model = MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet'
)

# "Đóng băng" các lớp của base_model để không huấn luyện lại chúng
base_model.trainable = False

# Xây dựng mô hình mới bằng cách thêm các lớp của chúng ta vào
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(), # Chuyển feature maps thành vector
    layers.Dropout(0.5),             # Lớp Dropout để giảm overfitting
    layers.Dense(17, activation='softmax') # Lớp output với 17 units (17 món ăn) và softmax
])

# Biên dịch mô hình (Compile)
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# In ra cấu trúc của mô hình
model.summary()
print("\n✅ Đã xây dựng và biên dịch mô hình thành công!")


# BƯỚC 3: HUẤN LUYỆN MÔ HÌNH

# Thiết lập số epochs để huấn luyện
EPOCHS = 20 # Bạn có thể bắt đầu với 20 và tăng lên nếu cần

print("\n🚀 Bắt đầu quá trình huấn luyện...")

history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=validation_generator
)

print("\n🎉 HOÀN TẤT HUẤN LUYỆN!")


# BƯỚC 4: LƯU MÔ HÌNH VÀ VẼ BIỂU ĐỒ KẾT QUẢ

# --- VUI LÒNG THAY ĐỔI ĐƯỜNG DẪN BÊN DƯỚI ---
# Đường dẫn để lưu mô hình đã huấn luyện
model_save_path = '/content/sample_data.h5'
# ----------------------------------------------------

model.save(model_save_path)
print(f"\n💾 Mô hình đã được lưu tại: {model_save_path}")

# Vẽ biểu đồ accuracy và loss
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(EPOCHS)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
